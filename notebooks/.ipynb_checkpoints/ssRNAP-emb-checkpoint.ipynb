````python
import random
from collections import Counter
from tqdm import tqdm

import torch
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import esm
````

````python
import scipy
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.decomposition import PCA
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.svm import SVC, SVR
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression, SGDRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import confusion_matrix
from scipy.spatial.distance import cosine, euclidean
````

## Add the path to your embeddings here:


````python
FASTA_PATH = "../data/ncbi-bacteriophage-human-meta.fasta"
EMB_PATH = "../output/ncbi-bacteriophage-human-meta_reprs" # Path to directory of embeddings
ECOL_EMB_PATH = "../output/UP000000625_83333_reprs"
POL_EMB_PATH = "../output/pol_reprs"

EMB_LAYER = 33
````

### Load T7 RNAP embedding

````python
t7_rnap_header = "sp|P00573|RPOL_BPT7 T7 RNA polymerase OS=Escherichia phage T7 OX=10760 GN=1 PE=1 SV=2"
fn = f'{POL_EMB_PATH}/{t7_rnap_header}.pt'
embs = torch.load(fn)
t7_rnap_repr = embs['mean_representations'][EMB_LAYER].numpy()
````

## Load embeddings

````python
rows_list = []

n = 0
for header, _seq in esm.data.read_fasta(FASTA_PATH):
    fn = f'{EMB_PATH}/{header}.pt'
    embs = torch.load(fn)
    mean_repr = embs['mean_representations'][EMB_LAYER].numpy()
    ID, desc = (header.split("|")[0], header.split("|")[1])
    # Distance calculations between T7 RNAP
    cos_distance = cosine(t7_rnap_repr, mean_repr)
    euc_distance = euclidean(t7_rnap_repr, mean_repr)
    rows_list.append({"ID": ID, 
                      "desc": desc, 
                      "cosine": cos_distance,
                      "euclidean": euc_distance,
                      "repr": mean_repr,})

df = pd.DataFrame(rows_list)
df
#Xs = torch.stack(Xs, dim=0).numpy()
#print(Xs.shape)
````

````python
df.sort_values('cosine').loc[:, ["ID", "desc", "cosine", "euclidean"]].to_csv("phage_emb_cos.csv")
df.sort_values('euclidean').loc[:, ["ID", "desc", "cosine", "euclidean"]].to_csv("phage_emb_euc.csv")
df.to_csv("phage_emb_full.csv")
````

### PCA

Principal Component Analysis is a popular technique for dimensionality reduction. Given `n_features` (1280 in our case), PCA computes a new set of `X` that "best explain the data." We've found that this enables downstream models to be trained faster with minimal loss in performance.  

Here, we set `X` to 60, but feel free to change it!


````python
num_pca_components = 60
pca = PCA(num_pca_components)
Xs_train_pca = pca.fit_transform(df.loc[:,"repr"].to_list())
````

````python
column_names = [f'PC{i+1}' for i in range(Xs_train_pca.shape[1])]
df_pca = pd.DataFrame(data=Xs_train_pca, columns=column_names)
df_pca["type"] = df["type"]
df_pca["euclidean"] = df["euclidean"]
````

<a id='viz_embeddings'></a>
## Visualize Embeddings

Here, we plot the first two principal components on the x- and y- axes. Each point is then colored by its scaled effect (what we want to predict).

Visually, we can see a separation based on color/effect, suggesting that our representations are useful for this task, without any task-specific training!

````python
df['type'] = np.where(df['desc'].str.contains("RNA poly", case=False), "RNA pol-like", "other")
df['type'] = np.where(df['desc'].str.contains("DNA poly", case=False), "DNA pol-like", df['type'])
````

````python
df_pca_top = df_pca.sort_values("euclidean").iloc[:500]
sns.scatterplot(data=df_pca_top, x="PC1", y="PC2", hue="type", s=50, hue_order=["DNA pol-like", "RNA pol-like", "other"])
````

````python
palette = ['#FF6B00', '#87CEEB', '#46505A']
hue_order = ["RNA pol-like", "DNA pol-like", "other"]

df_cos_sorted = df.sort_values('cosine').iloc[:50].reset_index(drop=True).reset_index()
df_euc_sorted = df.sort_values('euclidean').iloc[:50].reset_index(drop=True).reset_index()
fig, ax = plt.subplots(1,2)
sns.scatterplot(data=df_cos_sorted, x="cosine", y="index", ax=ax[0], hue="type", hue_order=hue_order, palette=palette)
sns.scatterplot(data=df_euc_sorted, x="euclidean", y="index", ax=ax[1], hue="type", hue_order=hue_order, palette=palette)
ax[0].set_ylabel("rank")
ax[1].set_ylabel("")
fig.show()
````

````python
palette = ['#FF6B00', '#87CEEB', '#46505A']
sns.boxplot(data = df.sort_values('euclidean').iloc[:500], x = "type", y = "euclidean",  order=["RNA pol-like", "DNA pol-like", "other"], palette=palette)
plt.ylabel("distance (top 500)")
plt.xlabel("")
````

````python
palette = ['#FF6B00', '#87CEEB', '#46505A']
sns.boxplot(data = df, x = "type", y = "euclidean", order=["RNA pol-like", "DNA pol-like", "other"], palette=palette)
plt.ylabel("distance (all ORFs)")
plt.xlabel("")
````

### Using embeddings from layer 0

````python
t7_rnap_header = "sp|P00573|RPOL_BPT7 T7 RNA polymerase OS=Escherichia phage T7 OX=10760 GN=1 PE=1 SV=2"
fn = f'{POL_EMB_PATH}/{t7_rnap_header}.pt'
embs = torch.load(fn)
t7_rnap_repr_layer_0 = embs['mean_representations'][0].numpy()
````

````python
rows_list = []

n = 0
for header, _seq in esm.data.read_fasta(FASTA_PATH):
    fn = f'{EMB_PATH}/{header}.pt'
    embs = torch.load(fn)
    mean_repr = embs['mean_representations'][0].numpy()
    ID, desc = (header.split("|")[0], header.split("|")[1])
    # Distance calculations between T7 RNAP
    cos_distance = cosine(t7_rnap_repr_layer_0, mean_repr)
    euc_distance = euclidean(t7_rnap_repr_layer_0, mean_repr)
    rows_list.append({"ID": ID, 
                      "desc": desc, 
                      "cosine": cos_distance,
                      "euclidean": euc_distance,
                      "repr": mean_repr,})

df_layer_0 = pd.DataFrame(rows_list)
df_layer_0
````

````python
df_layer_0['type'] = np.where(df_layer_0['desc'].str.contains("RNA poly", case=False), "RNA pol-like", "other")
df_layer_0['type'] = np.where(df_layer_0['desc'].str.contains("DNA poly", case=False), "DNA pol-like", df_layer_0['type'])
````

````python
palette = ['#46505A', '#87CEEB', '#FF6B00']
df_euc_sorted = df_layer_0.sort_values('euclidean').iloc[:100].reset_index(drop=True).reset_index()
fig, ax = plt.subplots(1,2,figsize=(9,5))
sns.scatterplot(data=df_euc_sorted, x="euclidean", y="index", ax=ax[0], hue="type", palette=palette)
sns.boxplot(data = df_layer_0.sort_values('euclidean').iloc[:500], x = "type", y = "euclidean", palette=palette, ax=ax[1])
ax[0].set_ylabel("rank")
ax[1].set_ylabel("distance (top 500)")
ax[1].set_xlabel("")
fig.show()
````
